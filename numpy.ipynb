```python
import numpy as np

arr2 = np.arange(16)
print("Shape of arr2:", arr2.shape)

arr2_copy = arr2.copy()
print("arr2:", arr2)
print("arr2_copy:", arr2_copy)

arr2 = arr2.reshape(4, 4)
print("Reshaped arr2:\n", arr2)
print("arr2_copy remains unchanged:", arr2_copy)

print("Number of dimensions in arr2:", arr2.ndim)

rows, cols = arr2.shape
print(f"arr2 has {rows} rows and {cols} columns")

print("Number of elements in arr2:", arr2.size)

print("Data type of arr2:", arr2.dtype)

print("50th percentile of arr2:", np.percentile(arr2, 50))

print("Median of arr2 elements:", np.median(arr2))
```

```python
import numpy as np
import matplotlib.pyplot as plt 

mean = 3
std_dev = 4
num_elements = 100

s = np.random.normal(mean,std_dev, num_elements)

plt.hist(s, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data")
plt.show()
```
